```python
import os
import glob
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

def set_seed(seed=42):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

class HARDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.long)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

def load_opportunity_local_data(data_path="/content/drive/MyDrive/OpportunityUCIDataset/dataset", window_size=30, step=3):
    print("Loading Opportunity dataset from local files...")

    target_activities = {
        406516: 'Open Door 1',
        406517: 'Open Door 2',
        404516: 'Close Door 1',
        404517: 'Close Door 2',
        406520: 'Open Fridge',
        404520: 'Close Fridge',
        406505: 'Open Dishwasher',
        404505: 'Close Dishwasher',
        406519: 'Open Drawer 1',
        404519: 'Close Drawer 1',
        406511: 'Open Drawer 2',
        404511: 'Close Drawer 2',
        406508: 'Open Drawer 3',
        404508: 'Close Drawer 3',
        408512: 'Clean Table',
        407521: 'Drink from Cup',
        405506: 'Toggle Switch'
    }

    try:
        import glob

        dat_files = glob.glob(os.path.join(data_path, "**", "*.dat"), recursive=True)

        if not dat_files:
            print(f"No .dat files found in {data_path}")
            return None, None, None

        print(f"Found {len(dat_files)} .dat files")

        all_data = []
        all_labels = []

        for file_path in dat_files:
            print(f"  Processing {os.path.basename(file_path)}...")
            try:
                data = pd.read_csv(file_path, sep='\s+', header=None, na_values='NaN')

                if data.empty or data.shape[1] < 100:
                    continue

                data = data.fillna(method='bfill').fillna(method='ffill')

                if data.shape[1] >= 100:
                    features = data.iloc[:, 1:78].values
                else:
                    features = data.iloc[:, 1:-5].values

                if data.shape[1] > 249:
                    labels = data.iloc[:, 249].values
                else:
                    labels = data.iloc[:, -1].values

                valid_mask = (np.isfinite(features).all(axis=1) &
                             np.isfinite(labels) &
                             ~np.isnan(labels))

                features = features[valid_mask]
                labels = labels[valid_mask]

                target_mask = np.isin(labels, list(target_activities.keys()))

                if not np.any(target_mask):
                    print(f"    No target activities found in {os.path.basename(file_path)}")
                    continue

                features = features[target_mask]
                labels = labels[target_mask]

                all_data.append(features)
                all_labels.extend(labels)

                print(f"    Found {len(labels)} samples, feature dim: {features.shape[1]}")

            except Exception as e:
                print(f"    Error processing {file_path}: {e}")
                continue

        if not all_data:
            print("No valid data loaded with target activities")
            return None, None, None

        X_combined = np.vstack(all_data)
        y_combined = np.array(all_labels)

        print(f"Combined data shape: {X_combined.shape}")
        print(f"Found activities: {np.unique(y_combined)}")

        unique_labels = np.unique(y_combined)
        label_mapping = {}
        activity_labels = []

        for idx, label in enumerate(unique_labels):
            if label in target_activities:
                label_mapping[label] = idx
                activity_labels.append(target_activities[label])

        y_encoded = np.array([label_mapping[label] for label in y_combined])

        print(f"Label mapping: {label_mapping}")
        print(f"Activity labels: {activity_labels}")

        scaler = MinMaxScaler()
        X_scaled = scaler.fit_transform(X_combined)

        if np.any(np.isnan(X_scaled)) or np.any(np.isinf(X_scaled)):
            print("Cleaning NaN/inf values after scaling...")
            X_scaled = np.nan_to_num(X_scaled, nan=0.0, posinf=1.0, neginf=-1.0)

        all_windows = []
        all_window_labels = []

        start = 0
        while start + window_size <= len(X_scaled):
            window_data = X_scaled[start:start + window_size, :]
            window_label = y_encoded[start + window_size - 1]

            if np.isfinite(window_data).all():
                all_windows.append(window_data)
                all_window_labels.append(window_label)

            start += step

        if not all_windows:
            print("No valid windows created")
            return None, None, None

        X_sequences = np.array(all_windows, dtype=np.float32)
        X_sequences = np.transpose(X_sequences, (0, 2, 1))
        y_sequences = np.array(all_window_labels)

        print(f"Final data validation:")
        print(f"  X range: [{X_sequences.min():.4f}, {X_sequences.max():.4f}]")
        print(f"  Contains NaN: {np.any(np.isnan(X_sequences))}")
        print(f"  Contains inf: {np.any(np.isinf(X_sequences))}")

        print(f"Opportunity - X_sequences shape: {X_sequences.shape}")
        print(f"Number of classes: {len(activity_labels)}")
        print(f"Class distribution: {np.bincount(y_sequences)}")

        return X_sequences, y_sequences, activity_labels

    except Exception as e:
        print(f"Error loading Opportunity dataset: {e}")
        return None, None, None

class CBAMBlock1D(nn.Module):
    def __init__(self, channels, reduction=4, kernel_size=7):
        super().__init__()
        self.pool_avg = nn.AdaptiveAvgPool1d(1)
        self.pool_max = nn.AdaptiveMaxPool1d(1)
        self.mlp = nn.Sequential(
            nn.Linear(channels, channels // reduction, bias=False),
            nn.ReLU(),
            nn.Linear(channels // reduction, channels, bias=False)
        )
        self.sigmoid_channel = nn.Sigmoid()
        self.conv_spatial = nn.Conv1d(2, 1, kernel_size=kernel_size, padding=kernel_size // 2)
        self.sigmoid_spatial = nn.Sigmoid()

    def forward(self, x):
        b, c, t = x.size()
        avg_out = self.mlp(self.pool_avg(x).view(b, c))
        max_out = self.mlp(self.pool_max(x).view(b, c))
        scale = self.sigmoid_channel(avg_out + max_out).view(b, c, 1)
        x = x * scale

        avg_spatial = torch.mean(x, dim=1, keepdim=True)
        max_spatial, _ = torch.max(x, dim=1, keepdim=True)
        spatial = torch.cat([avg_spatial, max_spatial], dim=1)
        spatial = self.sigmoid_spatial(self.conv_spatial(spatial))
        x = x * spatial
        return x

class LightMambaBlock1D(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.norm = nn.LayerNorm(dim)
        self.proj_in = nn.Linear(dim, dim)
        self.dw_conv = nn.Conv1d(dim, dim, kernel_size=3, padding=1, groups=dim)
        self.gate = nn.Sigmoid()
        self.proj_out = nn.Linear(dim, dim)

    def forward(self, x):
        x = self.norm(x)
        x_proj = self.proj_in(x)
        x_conv = self.dw_conv(x_proj.transpose(1, 2)).transpose(1, 2)
        x = self.proj_out(x_conv * self.gate(x_proj))
        return x

class MahalanobisContextPrototypes(nn.Module):
    def __init__(self, feature_dim, num_classes):
        super().__init__()
        self.feature_dim = feature_dim
        self.num_classes = num_classes

        self.prototype_centers = nn.Parameter(torch.randn(num_classes, feature_dim) * 0.1)

        self.log_cov_diag = nn.Parameter(torch.zeros(num_classes, feature_dim))

        self.context_center_shift = nn.Sequential(
            nn.Linear(feature_dim, feature_dim // 2),
            nn.ReLU(),
            nn.Linear(feature_dim // 2, feature_dim),
            nn.Tanh()
        )

        self.context_cov_scale = nn.Sequential(
            nn.Linear(feature_dim, feature_dim // 2),
            nn.ReLU(),
            nn.Linear(feature_dim // 2, feature_dim),
            nn.Sigmoid()
        )

        self.shift_scale = nn.Parameter(torch.tensor(0.1))
        self.cov_scale = nn.Parameter(torch.tensor(0.1))

    def compute_mahalanobis_distance(self, x, centers, cov_diag):
        batch_size = x.size(0)

        x_expanded = x.unsqueeze(1).expand(-1, self.num_classes, -1)
        centers_expanded = centers.unsqueeze(0).expand(batch_size, -1, -1)
        cov_diag_expanded = cov_diag.unsqueeze(0).expand(batch_size, -1, -1)

        diff = x_expanded - centers_expanded

        inv_var = 1.0 / (torch.exp(cov_diag_expanded) + 1e-6)

        mahal_dist = torch.sum(diff ** 2 * inv_var, dim=2)

        return mahal_dist

    def forward(self, x):
        batch_size = x.size(0)

        context_shift = self.context_center_shift(x)
        adjusted_centers = self.prototype_centers + self.shift_scale * context_shift.unsqueeze(1)
        adjusted_centers = adjusted_centers.view(-1, self.feature_dim)
        adjusted_centers = adjusted_centers.view(batch_size, self.num_classes, self.feature_dim)
        adjusted_centers = adjusted_centers.mean(dim=0)

        context_cov_scale = self.context_cov_scale(x)
        adjusted_cov_diag = self.log_cov_diag + self.cov_scale * context_cov_scale.unsqueeze(1).mean(dim=0)

        distances = self.compute_mahalanobis_distance(x, adjusted_centers, adjusted_cov_diag)

        logits = -distances

        return logits

class MambaOnly(nn.Module):
    def __init__(self, in_channels, num_classes, patch_size=8, hidden_dim=128, num_mamba_blocks=4):
        super().__init__()
        self.patch_size = patch_size
        self.token_dim = in_channels * patch_size

        self.down_proj = nn.Conv1d(self.token_dim, hidden_dim, kernel_size=1)
        self.mamba_blocks = nn.ModuleList([LightMambaBlock1D(hidden_dim) for _ in range(num_mamba_blocks)])
        self.up_proj = nn.Conv1d(hidden_dim, self.token_dim, kernel_size=1)

        self.global_pool = nn.AdaptiveAvgPool1d(1)
        self.classifier = nn.Linear(in_channels, num_classes)

        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.Conv1d):
                nn.init.xavier_uniform_(m.weight)

    def forward(self, x):
        b, c, t = x.size()
        assert t % self.patch_size == 0, f"T ({t}) must be divisible by patch_size ({self.patch_size})"

        x = x.view(b, c, t // self.patch_size, self.patch_size)
        x = x.permute(0, 2, 1, 3).reshape(b, t // self.patch_size, -1)

        x = self.down_proj(x.transpose(1, 2)).transpose(1, 2)
        for block in self.mamba_blocks:
            x = block(x) + x
        x = self.up_proj(x.transpose(1, 2)).transpose(1, 2)

        x = x.view(b, t // self.patch_size, c, self.patch_size).permute(0, 2, 1, 3)
        x = x.reshape(b, c, t)
        x = self.global_pool(x).squeeze(-1)
        return self.classifier(x)

class MambaCBAMPatch(nn.Module):
    def __init__(self, in_channels, num_classes, patch_size=8, hidden_dim=128, num_mamba_blocks=4):
        super().__init__()
        self.attn = CBAMBlock1D(in_channels)
        self.patch_size = patch_size
        self.token_dim = in_channels * patch_size

        self.down_proj = nn.Conv1d(self.token_dim, hidden_dim, kernel_size=1)
        self.mamba_blocks = nn.ModuleList([LightMambaBlock1D(hidden_dim) for _ in range(num_mamba_blocks)])
        self.up_proj = nn.Conv1d(hidden_dim, self.token_dim, kernel_size=1)

        self.global_pool = nn.AdaptiveAvgPool1d(1)
        self.classifier = nn.Linear(in_channels, num_classes)

        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.Conv1d):
                nn.init.xavier_uniform_(m.weight)

    def forward(self, x):
        b, c, t = x.size()
        assert t % self.patch_size == 0, f"T ({t}) must be divisible by patch_size ({self.patch_size})"

        x = self.attn(x)
        x = x.view(b, c, t // self.patch_size, self.patch_size)
        x = x.permute(0, 2, 1, 3).reshape(b, t // self.patch_size, -1)

        x = self.down_proj(x.transpose(1, 2)).transpose(1, 2)
        for block in self.mamba_blocks:
            x = block(x) + x
        x = self.up_proj(x.transpose(1, 2)).transpose(1, 2)

        x = x.view(b, t // self.patch_size, c, self.patch_size).permute(0, 2, 1, 3)
        x = x.reshape(b, c, t)
        x = self.global_pool(x).squeeze(-1)
        return self.classifier(x)

class MahalanobisCBAM_Mamba_HAR(nn.Module):
    def __init__(self, in_channels, num_classes, patch_size=8, hidden_dim=128, num_mamba_blocks=4):
        super().__init__()
        self.attn = CBAMBlock1D(in_channels)
        self.patch_size = patch_size
        self.token_dim = in_channels * patch_size

        self.down_proj = nn.Conv1d(self.token_dim, hidden_dim, kernel_size=1)
        self.mamba_blocks = nn.ModuleList([LightMambaBlock1D(hidden_dim) for _ in range(num_mamba_blocks)])
        self.up_proj = nn.Conv1d(hidden_dim, self.token_dim, kernel_size=1)

        self.global_pool = nn.AdaptiveAvgPool1d(1)
        self.classifier = MahalanobisContextPrototypes(feature_dim=in_channels, num_classes=num_classes)

        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.Conv1d):
                nn.init.xavier_uniform_(m.weight)

    def forward(self, x):
        b, c, t = x.size()
        assert t % self.patch_size == 0, f"T ({t}) must be divisible by patch_size ({self.patch_size})"

        x = self.attn(x)
        x = x.view(b, c, t // self.patch_size, self.patch_size)
        x = x.permute(0, 2, 1, 3).reshape(b, t // self.patch_size, -1)

        x = self.down_proj(x.transpose(1, 2)).transpose(1, 2)
        for block in self.mamba_blocks:
            x = block(x) + x
        x = self.up_proj(x.transpose(1, 2)).transpose(1, 2)

        x = x.view(b, t // self.patch_size, c, self.patch_size).permute(0, 2, 1, 3)
        x = x.reshape(b, c, t)
        x = self.global_pool(x).squeeze(-1)
        return self.classifier(x)

def get_patch_size_for_dataset(dataset_name, window_size):
    if dataset_name == 'Opportunity':
        return 5
    else:
        return 8

def calculate_sensitivity_specificity(y_true, y_pred, num_classes):
    cm = confusion_matrix(y_true, y_pred, labels=range(num_classes))

    sensitivity = []
    specificity = []

    for i in range(num_classes):
        tp = cm[i, i]
        fn = np.sum(cm[i, :]) - tp
        fp = np.sum(cm[:, i]) - tp
        tn = np.sum(cm) - tp - fn - fp

        sens = tp / (tp + fn) if (tp + fn) > 0 else 0
        spec = tn / (tn + fp) if (tn + fp) > 0 else 0

        sensitivity.append(sens)
        specificity.append(spec)

    return np.mean(sensitivity), np.mean(specificity)

def train_model(model, train_loader, val_loader, epochs=50, lr=0.001, device='cuda'):
    model.to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.CrossEntropyLoss()
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=10)

    best_val_acc = 0
    patience_counter = 0
    patience = 20

    train_losses = []
    train_accs = []
    val_losses = []
    val_accs = []

    for epoch in range(epochs):
        model.train()
        total_loss = 0
        correct = 0
        total = 0

        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            pred = model(xb)
            loss = criterion(pred, yb)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
            _, predicted = pred.max(1)
            total += yb.size(0)
            correct += predicted.eq(yb).sum().item()

        train_loss = total_loss / len(train_loader)
        train_acc = correct / total
        train_losses.append(train_loss)
        train_accs.append(train_acc)

        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                pred = model(xb)
                loss = criterion(pred, yb)

                val_loss += loss.item()
                _, predicted = pred.max(1)
                val_total += yb.size(0)
                val_correct += predicted.eq(yb).sum().item()

        val_loss = val_loss / len(val_loader)
        val_acc = val_correct / val_total
        val_losses.append(val_loss)
        val_accs.append(val_acc)

        scheduler.step(val_loss)

        print(f"Epoch {epoch+1}/{epochs} - Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            patience_counter = 0
        else:
            patience_counter += 1

        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch+1}")
            break

    return train_losses, train_accs, val_losses, val_accs

def evaluate_model(model, test_loader, num_classes, device='cuda'):
    model.eval()
    all_preds = []
    all_labels = []
    all_features = []

    with torch.no_grad():
        for xb, yb in test_loader:
            xb, yb = xb.to(device), yb.to(device)
            pred = model(xb)

            if hasattr(model, 'global_pool'):
                x = xb
                if hasattr(model, 'attn'):
                    x = model.attn(x)

                b, c, t = x.size()
                if hasattr(model, 'patch_size'):
                    x = x.view(b, c, t // model.patch_size, model.patch_size)
                    x = x.permute(0, 2, 1, 3).reshape(b, t // model.patch_size, -1)
                    x = model.down_proj(x.transpose(1, 2)).transpose(1, 2)
                    for block in model.mamba_blocks:
                        x = block(x) + x
                    x = model.up_proj(x.transpose(1, 2)).transpose(1, 2)
                    x = x.view(b, t // model.patch_size, c, model.patch_size).permute(0, 2, 1, 3)
                    x = x.reshape(b, c, t)
                    features = model.global_pool(x).squeeze(-1)
                else:
                    features = model.global_pool(x).squeeze(-1)

                all_features.extend(features.cpu().numpy())

            preds = pred.argmax(dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(yb.cpu().numpy())

    acc = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='weighted')
    precision = precision_score(all_labels, all_preds, average='weighted')
    recall = recall_score(all_labels, all_preds, average='weighted')
    sensitivity, specificity = calculate_sensitivity_specificity(all_labels, all_preds, num_classes)
    cm = confusion_matrix(all_labels, all_preds)
    cm_normalized = confusion_matrix(all_labels, all_preds, normalize='true')

    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall,
        'sensitivity': sensitivity,
        'specificity': specificity,
        'confusion_matrix': cm,
        'confusion_matrix_normalized': cm_normalized,
        'features': np.array(all_features) if all_features else None,
        'labels': np.array(all_labels)
    }

def visualize_tsne(features_dict, labels, activity_labels, title_prefix=""):
    for model_name, features in features_dict.items():
        if features is not None:
            plt.figure(figsize=(8, 6))

            tsne = TSNE(n_components=2, random_state=42, perplexity=30)
            features_2d = tsne.fit_transform(features)

            unique_labels = np.unique(labels)
            colors = plt.cm.tab10(np.linspace(0, 1, len(unique_labels)))

            for label, color in zip(unique_labels, colors):
                mask = labels == label
                if label < len(activity_labels):
                    label_name = activity_labels[label]
                else:
                    label_name = f"Class {label}"
                plt.scatter(features_2d[mask, 0], features_2d[mask, 1],
                           c=[color], label=label_name, alpha=0.7, s=20)

            plt.title(f'{title_prefix}{model_name}')
            plt.xlabel('t-SNE 1')
            plt.ylabel('t-SNE 2')
            plt.legend()
            plt.tight_layout()
            plt.show()

def run_cross_validation(X, y, num_classes, dataset_name, activity_labels):
    print(f"\n=== 5-Fold Cross Validation for {dataset_name} ===")

    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    results = []

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    window_size = X.shape[2]
    patch_size = get_patch_size_for_dataset(dataset_name, window_size)
    print(f"Using patch_size: {patch_size} for window_size: {window_size}")

    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y)):
        print(f"\nFold {fold+1}/5")

        X_train_fold, X_test_fold = X[train_idx], X[test_idx]
        y_train_fold, y_test_fold = y[train_idx], y[test_idx]

        X_train, X_val, y_train, y_val = train_test_split(
            X_train_fold, y_train_fold, test_size=0.2, stratify=y_train_fold, random_state=42
        )

        train_loader = DataLoader(HARDataset(X_train, y_train), batch_size=64, shuffle=True)
        val_loader = DataLoader(HARDataset(X_val, y_val), batch_size=64)
        test_loader = DataLoader(HARDataset(X_test_fold, y_test_fold), batch_size=64)

        in_channels = X.shape[1]
        model = MahalanobisCBAM_Mamba_HAR(in_channels=in_channels, num_classes=num_classes, patch_size=patch_size)

        train_losses, train_accs, val_losses, val_accs = train_model(
            model, train_loader, val_loader, epochs=50, lr=0.001, device=device
        )

        metrics = evaluate_model(model, test_loader, num_classes, device)

        fold_result = {
            'dataset': dataset_name,
            'fold': fold + 1,
            'accuracy': metrics['accuracy'],
            'f1': metrics['f1'],
            'precision': metrics['precision'],
            'recall': metrics['recall'],
            'sensitivity': metrics['sensitivity'],
            'specificity': metrics['specificity']
        }

        results.append(fold_result)

        print(f"Fold {fold+1} Results:")
        print(f"Accuracy: {metrics['accuracy']:.4f}")
        print(f"F1: {metrics['f1']:.4f}")
        print(f"Precision: {metrics['precision']:.4f}")
        print(f"Recall: {metrics['recall']:.4f}")
        print(f"Sensitivity: {metrics['sensitivity']:.4f}")
        print(f"Specificity: {metrics['specificity']:.4f}")
        print("Class Mapping:")
        for i, label in enumerate(activity_labels):
            if i < num_classes:
                print(f"  Class {i}: {label}")

        class_names = [activity_labels[i] if i < len(activity_labels) else f"Class_{i}" for i in range(num_classes)]

        print("Confusion Matrix (Count):")
        cm_count_df = pd.DataFrame(metrics['confusion_matrix'],
                                  index=[f"True_{name}" for name in class_names],
                                  columns=[f"Pred_{name}" for name in class_names])
        display(cm_count_df)

        print("Confusion Matrix (Normalized):")
        cm_norm_df = pd.DataFrame(metrics['confusion_matrix_normalized'],
                                 index=[f"True_{name}" for name in class_names],
                                 columns=[f"Pred_{name}" for name in class_names])
        cm_norm_df = (cm_norm_df * 100).astype(int) / 100
        display(cm_norm_df)

    avg_results = {
        'dataset': dataset_name,
        'fold': 'Average',
        'accuracy': np.mean([r['accuracy'] for r in results]),
        'f1': np.mean([r['f1'] for r in results]),
        'precision': np.mean([r['precision'] for r in results]),
        'recall': np.mean([r['recall'] for r in results]),
        'sensitivity': np.mean([r['sensitivity'] for r in results]),
        'specificity': np.mean([r['specificity'] for r in results]),
    }

    results.append(avg_results)

    print(f"\n=== Average Results for {dataset_name} ===")
    print(f"Accuracy: {avg_results['accuracy']:.4f}")
    print(f"F1: {avg_results['f1']:.4f}")
    print(f"Precision: {avg_results['precision']:.4f}")
    print(f"Recall: {avg_results['recall']:.4f}")
    print(f"Sensitivity: {avg_results['sensitivity']:.4f}")
    print(f"Specificity: {avg_results['specificity']:.4f}")

    return results

def run_ablation_study(X, y, num_classes, dataset_name, activity_labels):
    print(f"\n=== Ablation Study for {dataset_name} ===")

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    window_size = X.shape[2]
    patch_size = get_patch_size_for_dataset(dataset_name, window_size)
    print(f"Using patch_size: {patch_size} for window_size: {window_size}")

    X_temp, X_test, y_temp, y_test = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=42
    )
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=0.25, stratify=y_temp, random_state=42
    )

    train_loader = DataLoader(HARDataset(X_train, y_train), batch_size=64, shuffle=True)
    val_loader = DataLoader(HARDataset(X_val, y_val), batch_size=64)
    test_loader = DataLoader(HARDataset(X_test, y_test), batch_size=64)

    in_channels = X.shape[1]

    models = {
        'Mamba Only': MambaOnly(in_channels=in_channels, num_classes=num_classes, patch_size=patch_size),
        'Mamba + CBAM + Patch': MambaCBAMPatch(in_channels=in_channels, num_classes=num_classes, patch_size=patch_size),
        'Mamba + CBAM + Patch + Mahalanobis Context Aware': MahalanobisCBAM_Mamba_HAR(in_channels=in_channels, num_classes=num_classes, patch_size=patch_size)
    }

    results = {}
    features_dict = {}
    ablation_results = []

    for model_name, model in models.items():
        print(f"\nTraining {model_name}...")

        train_losses, train_accs, val_losses, val_accs = train_model(
            model, train_loader, val_loader, epochs=50, lr=0.001, device=device
        )

        metrics = evaluate_model(model, test_loader, num_classes, device)

        results[model_name] = metrics
        features_dict[model_name] = metrics['features']

        ablation_entry = {
            'dataset': dataset_name,
            'model': model_name,
            'accuracy': metrics['accuracy'],
            'f1': metrics['f1'],
            'precision': metrics['precision'],
            'recall': metrics['recall'],
            'sensitivity': metrics['sensitivity'],
            'specificity': metrics['specificity']
        }
        ablation_results.append(ablation_entry)

        print(f"{model_name} Results:")
        print(f"Accuracy: {metrics['accuracy']:.4f}")
        print(f"F1: {metrics['f1']:.4f}")
        print(f"Precision: {metrics['precision']:.4f}")
        print(f"Recall: {metrics['recall']:.4f}")
        print(f"Sensitivity: {metrics['sensitivity']:.4f}")
        print(f"Specificity: {metrics['specificity']:.4f}")
        print("Class Mapping:")
        for i, label in enumerate(activity_labels):
            if i < num_classes:
                print(f"  Class {i}: {label}")

        class_names = [activity_labels[i] if i < len(activity_labels) else f"Class_{i}" for i in range(num_classes)]

        print("Confusion Matrix (Count):")
        cm_count_df = pd.DataFrame(metrics['confusion_matrix'],
                                  index=[f"True_{name}" for name in class_names],
                                  columns=[f"Pred_{name}" for name in class_names])
        display(cm_count_df)

        print("Confusion Matrix (Normalized):")
        cm_norm_df = pd.DataFrame(metrics['confusion_matrix_normalized'],
                                 index=[f"True_{name}" for name in class_names],
                                 columns=[f"Pred_{name}" for name in class_names])
        cm_norm_df = (cm_norm_df * 100).astype(int) / 100
        display(cm_norm_df)

    if any(f is not None for f in features_dict.values()):
        print(f"\nGenerating t-SNE visualization for {dataset_name}...")
        visualize_tsne(features_dict, y_test, activity_labels, title_prefix=f"{dataset_name} - ")

    return results, ablation_results

def main():
    all_cv_results = []
    all_ablation_results = []

    print(f"\n{'='*50}")
    print(f"Processing Opportunity Dataset - Target Activities Only")
    print(f"{'='*50}")

    try:
        X, y, activity_labels = load_opportunity_local_data()

        if X is not None and y is not None:
            num_classes = len(np.unique(y))
            print(f"Dataset shape: {X.shape}")
            print(f"Number of classes: {num_classes}")
            print(f"Activity labels: {activity_labels}")

            unique, counts = np.unique(y, return_counts=True)
            for i, (label_idx, count) in enumerate(zip(unique, counts)):
                print(f"  {activity_labels[label_idx]}: {count} samples")

            cv_results = run_cross_validation(X, y, num_classes, 'Opportunity', activity_labels)
            all_cv_results.extend(cv_results)

            results_dict, ablation_results = run_ablation_study(X, y, num_classes, 'Opportunity', activity_labels)
            all_ablation_results.extend(ablation_results)

            print("Opportunity dataset processing completed!")

        else:
            print("Failed to load Opportunity dataset")

    except Exception as e:
        print(f"Error processing Opportunity: {e}")

    print("\n" + "="*80)
    print("CREATING FINAL RESULTS DATAFRAME")
    print("="*80)

    try:
        if all_cv_results:
            cv_df = pd.DataFrame(all_cv_results)
            cv_df['experiment_type'] = 'Cross Validation'
            cv_df['model'] = 'MahalanobisCBAM_Mamba_HAR'
            cv_df = cv_df[['dataset', 'experiment_type', 'model', 'fold', 'accuracy', 'f1', 'precision', 'recall', 'sensitivity', 'specificity']]
        else:
            cv_df = pd.DataFrame(columns=['dataset', 'experiment_type', 'model', 'fold', 'accuracy', 'f1', 'precision', 'recall', 'sensitivity', 'specificity'])

        if all_ablation_results:
            ablation_df = pd.DataFrame(all_ablation_results)
            ablation_df['experiment_type'] = 'Ablation Study'
            ablation_df['fold'] = 'Single Run'
            ablation_df = ablation_df[['dataset', 'experiment_type', 'model', 'fold', 'accuracy', 'f1', 'precision', 'recall', 'sensitivity', 'specificity']]
        else:
            ablation_df = pd.DataFrame(columns=['dataset', 'experiment_type', 'model', 'fold', 'accuracy', 'f1', 'precision', 'recall', 'sensitivity', 'specificity'])

        if not cv_df.empty or not ablation_df.empty:
            final_df = pd.concat([cv_df, ablation_df], ignore_index=True)
        else:
            final_df = pd.DataFrame(columns=['dataset', 'experiment_type', 'model', 'fold', 'accuracy', 'f1', 'precision', 'recall', 'sensitivity', 'specificity'])

        print("\n" + "="*80)
        print("FINAL RESULTS - ALL EXPERIMENTS")
        print("="*80)
        display(final_df)

        print("\n" + "="*80)
        print("ABLATION STUDY SUMMARY")
        print("="*80)

        if not final_df.empty:
            for dataset_name in final_df['dataset'].unique():
                print(f"\n{dataset_name}:")
                dataset_ablation = final_df[(final_df['dataset'] == dataset_name) & (final_df['experiment_type'] == 'Ablation Study')]
                for _, row in dataset_ablation.iterrows():
                    print(f"  {row['model']}: Acc={row['accuracy']:.4f}, F1={row['f1']:.4f}")
        else:
            print("No results to display")

    except Exception as e:
        print(f"Error creating final DataFrame: {e}")

if __name__ == "__main__":
    main()
```

```python
def load_uci_har_data(dataset_path="/content/", window_size=128, step=64):
    print("Loading UCI HAR dataset...")

    SIGNALS = ["body_acc_x", "body_acc_y", "body_acc_z",
               "body_gyro_x", "body_gyro_y", "body_gyro_z",
               "total_acc_x", "total_acc_y", "total_acc_z"]

    X_list = []

    for signal in SIGNALS:
        train_file = os.path.join(dataset_path, f"{signal}_train.txt")
        test_file = os.path.join(dataset_path, f"{signal}_test.txt")

        try:
            train_signal = np.loadtxt(train_file, dtype=np.float32)
            test_signal = np.loadtxt(test_file, dtype=np.float32)
            combined_signal = np.concatenate([train_signal, test_signal], axis=0)
            X_list.append(combined_signal)
        except FileNotFoundError as e:
            print(f"Error loading file: {e}")
            return None, None, None

    if not X_list:
        print("No signal data loaded")
        return None, None, None

    X = np.stack(X_list, axis=-1)
    X = np.transpose(X, (0, 2, 1))

    # Handle NaNs
    X_df = pd.DataFrame(X.reshape(X.shape[0], -1))
    X_df = X_df.fillna(method='ffill').fillna(method='bfill')
    X = X_df.values.reshape(X.shape)

    try:
        y_train = np.loadtxt(os.path.join(dataset_path, "y_train.txt"), dtype=int)
        y_test = np.loadtxt(os.path.join(dataset_path, "y_test.txt"), dtype=int)
        y = np.concatenate([y_train, y_test], axis=0)
        y = y - 1

        label_encoder = LabelEncoder()
        y_encoded = label_encoder.fit_transform(y)
    except FileNotFoundError as e:
        print(f"Error loading label file: {e}")
        return None, None, None

    scaler = MinMaxScaler()
    X_flat = X.reshape(X.shape[0], -1)
    X_flat = scaler.fit_transform(X_flat)
    X = X_flat.reshape(X.shape)

    ucihar_activity_mapping = {
        0: 'Walking', 1: 'Walking Upstairs', 2: 'Walking Downstairs',
        3: 'Sitting', 4: 'Standing', 5: 'Laying'
    }

    activity_labels = []
    unique_labels = sorted(np.unique(y_encoded))
    for label in unique_labels:
        if label in ucihar_activity_mapping:
            activity_labels.append(ucihar_activity_mapping[label])
        else:
            activity_labels.append(f"Unknown_Activity_{label}")

    print(f"UCI HAR - X_sequences shape: {X.shape}")
    return X, y_encoded, activity_labels

def get_patch_size_for_dataset(dataset_name, window_size):
    if dataset_name == 'UCI-HAR':
        return 8
    elif dataset_name == 'Opportunity':
        return 5
    else:
        return 8

all_cv_results = []
all_ablation_results = []

print(f"\n{'='*50}")
print(f"Processing UCI-HAR Dataset")
print(f"{'='*50}")

try:
    X, y, activity_labels = load_uci_har_data()

    if X is not None and y is not None:
        num_classes = len(np.unique(y))
        print(f"Dataset shape: {X.shape}")
        print(f"Number of classes: {num_classes}")
        print(f"Activity labels: {activity_labels}")

        unique, counts = np.unique(y, return_counts=True)
        for i, (label_idx, count) in enumerate(zip(unique, counts)):
            print(f"  {activity_labels[label_idx]}: {count} samples")

        cv_results = run_cross_validation(X, y, num_classes, 'UCI-HAR', activity_labels)
        all_cv_results.extend(cv_results)

        results_dict, ablation_results = run_ablation_study(X, y, num_classes, 'UCI-HAR', activity_labels)
        all_ablation_results.extend(ablation_results)

        print("UCI-HAR dataset processing completed!")

    else:
        print("Failed to load UCI-HAR dataset")

except Exception as e:
    print(f"Error processing UCI-HAR: {e}")

print("\n" + "="*80)
print("CREATING FINAL RESULTS DATAFRAME")
print("="*80)

try:
    if all_cv_results:
        cv_df = pd.DataFrame(all_cv_results)
        cv_df['experiment_type'] = 'Cross Validation'
        cv_df['model'] = 'MahalanobisCBAM_Mamba_HAR'
        cv_df = cv_df[['dataset', 'experiment_type', 'model', 'fold', 'accuracy', 'f1', 'precision', 'recall', 'sensitivity', 'specificity']]
    else:
        cv_df = pd.DataFrame(columns=['dataset', 'experiment_type', 'model', 'fold', 'accuracy', 'f1', 'precision', 'recall', 'sensitivity', 'specificity'])

    if all_ablation_results:
        ablation_df = pd.DataFrame(all_ablation_results)
        ablation_df['experiment_type'] = 'Ablation Study'
        ablation_df['fold'] = 'Single Run'
        ablation_df = ablation_df[['dataset', 'experiment_type', 'model', 'fold', 'accuracy', 'f1', 'precision', 'recall', 'sensitivity', 'specificity']]
    else:
        ablation_df = pd.DataFrame(columns=['dataset', 'experiment_type', 'model', 'fold', 'accuracy', 'f1', 'precision', 'recall', 'sensitivity', 'specificity'])

    if not cv_df.empty or not ablation_df.empty:
        final_df = pd.concat([cv_df, ablation_df], ignore_index=True)
    else:
        final_df = pd.DataFrame(columns=['dataset', 'experiment_type', 'model', 'fold', 'accuracy', 'f1', 'precision', 'recall', 'sensitivity', 'specificity'])

    print("\n" + "="*80)
    print("FINAL RESULTS - UCI-HAR")
    print("="*80)
    display(final_df)

    print("\n" + "="*80)
    print("ABLATION STUDY SUMMARY")
    print("="*80)

    if not final_df.empty:
        for dataset_name in final_df['dataset'].unique():
            print(f"\n{dataset_name}:")
            dataset_ablation = final_df[(final_df['dataset'] == dataset_name) & (final_df['experiment_type'] == 'Ablation Study')]
            for _, row in dataset_ablation.iterrows():
                print(f"  {row['model']}: Acc={row['accuracy']:.4f}, F1={row['f1']:.4f}")
    else:
        print("No results to display")

except Exception as e:
    print(f"Error creating final DataFrame: {e}")
```
